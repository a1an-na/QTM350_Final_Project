```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the dataset
df = pd.read_csv('../data/population_dynamics_clean.csv')
```

```python
# Life Expectancy Rate by Country Over Time

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df,
    x="year",
    y="life_expectancy",
    hue="country",
    marker="o"
)

plt.title("Life Expectancy Over Time by Country", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Life Expectancy (Years)")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

```python
# Bar plot of average life expectancy per country
avg_life_exp = df.groupby("country")["life_expectancy"].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_life_exp, x="country", y="life_expectancy")
plt.title("Average Life Expectancy (20002023) by Country")
plt.ylabel("Average Life Expectancy (Years)")
plt.xlabel("Country")
plt.tight_layout()
plt.show()
```
